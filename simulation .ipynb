# 2. 
If you flip 8 coins, what is the probability of getting exactly 3 heads? What is the probability of getting more than 3 heads?

```python
trails= 1000
coins= 8 
flips =np.random.choice(["heads", "Tails"], size =(trails, coins))
flips 
```

```python
flips = pd.DataFrame(flips, columns=["first_coin", "second_coin","third_coin", "fourth_coin", "fifth_coin", "sixth_coin", "seventh_coin","eighth_coin"])
flips.head(20)
```

```python
def count_heads(row):
   return row.first_coin.count("heads") + row.second_coin.count("heads")+row.third_coin.count("heads")+row.fourth_coin.count("heads")+row.fifth_coin.count("heads")+row.sixth_coin.count("heads")+row.seventh_coin.count("heads")+row.eighth_coin.count("heads")
```

```python
flips["head_count"] = flips.apply(count_heads, axis=1)
flips
```

```python

# What are the chances of flipping exactly three heads?
flips[(flips.head_count == 3)]
```

```python
(flips.head_count == 3).mean()
```

```python
# more than 3 heads 
flips[(flips.head_count > 3)]
```

```python
(flips.head_count>3).mean()
```

# 3.
There are approximitely 3 web development cohorts for every 1 data science cohort at Codeup. Assuming that Codeup randomly selects an alumni to put on a billboard, what are the odds that the two billboards I drive past both have data science students on them?

```python
trails_stu= 100000
alumni=2
bill_board= np.random.choice (["web_dev","web_dev","web_dev","data_sci"], size = (trails_stu, alumni))
bill_board
```

```python
bill_board= pd.DataFrame(bill_board, columns= ["first1","second1"])
bill_board.head(10)

```

```python
def count_data(row):
   return row.first1.count("data_sci") + row.second1.count("data_sci")
```

```python

# Let's setup a 5rd column to count the number of data scientists  
bill_board["count_scientists"] = bill_board.apply(count_data, axis=1)
bill_board
```

```python
#odds have 2 data scientists 
bill_board[(bill_board.count_scientists == 2)]
```

```python
(bill_board.count_scientists == 2).mean()
```

 - What are the odds that after having 50 students download anaconda, no one has an installation issue? 100 students?

- What is the probability that we observe an installation issue within the first 150 students that download anaconda?
- How likely is it that 450 students all download anaconda without an issue?

```python
fails = np.random.choice(['fails', 'success'], size = (100_000, 50), p = (1/250, 249/250))
```

```python
((fails == 'success').sum(axis=1) == 50).mean() # probability of 50 successes
```

# 8
If 23 people are in the same room, what are the odds that two of them share a birthday? What if it's 20 people? 40?

```python

birthday = np.random.choice(range(365),size = (100_000, 23)) # first we are getting 23 samples from the size of 365

df = pd.DataFrame(birthday)
df['n_unique']=df.nunique(axis = 1)
df.head()
```

```python
# probability of all of them having a unique birthday
(df.n_unique == 23).mean()

# # finding the probability of people sharing birthday

(df.n_unique != 23).mean()
```

```python
# for 20 people

birthday = np.random.choice(range(365),size = (100_000, 20)) # first we are getting 20 samples from the size of 365

df = pd.DataFrame(birthday)
df['n_unique']=df.nunique(axis = 1)
df.head()
(df.n_unique != 20).mean()
```

```python

# for 40 people
birthday = np.random.choice(range(365),size = (100_000, 40)) # first we are getting 40 samples from the size of 365

df = pd.DataFrame(birthday)
df['n_unique']=df.nunique(axis = 1)
df.head()
(df.n_unique != 40).mean()
```
